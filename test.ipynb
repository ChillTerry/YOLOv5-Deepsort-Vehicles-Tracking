```python
# 用于获取画面的像素点坐标
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib qt5

# 加载截图
img = cv2.imread('./videos/snap.jpg') 

# 缩放
img = cv2.resize(img, (562,1000))

# 弹出窗口可以选择
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
t_l = (196,129)
t_r = (337,111)

b_l = (12,513)
b_r = (530,516)
```

```python
# 绘制，检查
cv2.circle(img,t_l,10,(0,0,255),-1)
cv2.circle(img,t_r,10,(0,0,255),-1)
cv2.circle(img,b_l,10,(0,0,255),-1)
cv2.circle(img,b_r,10,(0,0,255),-1)
```

```python
%matplotlib inline
plt.rcParams['figure.dpi'] = 200
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
